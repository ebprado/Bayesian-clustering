```python
## Controlando erros no cálculo do PSI via Teorema Central do Limite e método Delta

Defina

$$
\begin{aligned}
\text{PSI}(p, q) &= \sum_{j=1}^{k} (p_j - q_j) \log\left(\frac{p_j}{q_j}\right) \\
&= \sum_{j=1}^{k} (p_j - q_j) \left[ \log(p_j) - \log(q_j) \right]
\end{aligned}
$$

Usando o Teorema Central do Limite (TCL) e a aproximação da distribuição binomial pela distribuição normal, sabemos que para uma proporção $\widehat{p}_j$ obtida a partir de uma amostra de tamanho $n$:

$$
\widehat{p}_j \sim N\left(p_j, \frac{p_j (1 - p_j)}{n} \right)
$$

Também sabemos via Método Delta que se uma sequência de variáveis aleatórias $X_1, \ldots, X_n$ satisfaz:

$$
\sqrt{n}[X_n - \mu] \xrightarrow{d} N(0, \sigma^{2}),
$$

onde $\mathbb{E}(\bar{X}) = \mu \in \mathbb{R}$ e $\text{Var}(\bar{X}) = \sigma^{2} \in \mathbb{R}^{+}_{*}$, então para qualquer função derivável $g(\cdot)$ tal que $g'(\mu) \ne 0$:

$$
\sqrt{n}[g(\widehat{p}_j) - g(p_j)] \xrightarrow{d} N\left(0, \frac{p_j (1 - p_j)}{n} [g'(p_j)]^2 \right)
$$

O resultado acima nos ajuda a encontrar a distribuição de $\log(p_j)$, a qual é dada por:

$$
\log(\widehat{p}_j) \sim N\left(\log(p_j), \frac{1 - p_j}{n p_j}\right)
$$

Uma vez que o PSI é uma função de diferenças entre proporções e o logaritmo neperiano da razão entre as mesmas, precisamos encontrar as distribuições assintóticas de $p_j - q_j$ e $\log(p_j) - \log(q_j)$. 

Como todas variáveis aleatórias são normalmente distribuídas com diferentes médias e variâncias:

$$
\widehat{p}_j - \widehat{q}_j \sim N\left( p_j - q_j, \frac{p_j(1-p_j)}{n} + \frac{q_j(1-q_j)}{m} \right)
$$

e

$$
\log(\widehat{p}_j) - \log(\widehat{q}_j) \sim N\left( \log(p_j) - \log(q_j), \frac{1-p_j}{n p_j} + \frac{1-q_j}{m q_j} \right)
$$

Agora que sabemos as distribuições assintóticas de $p_j - q_j$ e $\log(p_j) - \log(q_j)$, podemos nos concentrar em encontrar a esperança e a variância do PSI. A ideia é construirmos um intervalo de confiança com o objetivo de encontrar tamanhos "ótimos" de amostras $n$ e $m$.

Defina $X_j = p_j - q_j$ e $Y_j = \log(p_j) - \log(q_j)$. Assumindo que $p$ e $q$ são independentes, então $X_j$ e $Y_j$ são independentes. Com isso:

$$
\mathbb{E}(X_jY_j) = (p_j - q_j) \left[ \log(p_j) - \log(q_j) \right]
$$

e

$$
\begin{aligned}
\text{Var}(X_jY_j) &= \mathbb{E}\left([X_jY_j]^2\right) - \left[\mathbb{E}(X_jY_j)\right]^2 \\
&= \mathbb{E}(X_j^2)\mathbb{E}(Y_j^2) - \left[\mathbb{E}(X_j)\mathbb{E}(Y_j)\right]^2 \\
&= (\sigma^{2}_{xj} + \mu_{xj}^{2}) (\sigma^{2}_{yj} + \mu_{yj}^{2}) - \mu_{xj}^{2} \mu_{yj}^{2} \\
&= \sigma^{2}_{xj} \sigma^{2}_{yj} + \sigma^{2}_{xj} \mu_{yj}^{2} + \mu_{xj}^{2} \sigma^{2}_{yj}
\end{aligned}
$$

onde:

$$
\mu_{xj} = p_j - q_j
$$

$$
\mu_{yj} = \log\left(\frac{p_j}{q_j}\right)
$$

$$
\sigma^{2}_{xj} = \frac{p_j(1-p_j)}{n} + \frac{q_j(1-q_j)}{m}
$$ 

$$
\sigma^{2}_{yj} = \frac{1-p_j}{n p_j} + \frac{1-q_j}{m q_j}
$$

Definindo $\text{Var}(\widehat{\text{PSI}}) = \sum_{j=1}^{k} \text{Var}(X_jY_j)$, o intervalo de confiança para o PSI é dado por:

$$
\widehat{\text{PSI}} \pm Z_{1 - \alpha/2} \sqrt{\text{Var}(\widehat{\text{PSI}})}
$$

Assumindo que $n = m$ e um erro amostral $\epsilon > 0$, temos que o tamanho amostral ótimo é:

$$
n = \frac{Z_{1 - \alpha/2}^{2} \times \text{Var}(\widehat{\text{PSI}})}{\epsilon^{2}}
$$
```

## Controlando erros no cálculo do PSI via Teorema Central do Limite e método Delta
Defina
$$
\begin{align}
\text{PSI}(p, q) & = \sum_{j=1}^{k} (p_j - q_j) \times \log(p_j / q_j) \nonumber \\
          & = \sum_{j=1}^{k} (p_j - q_j) \times [\log(p_j)  - \log(q_j)]. \nonumber
\end{align}
$$
Usando o Teorema Central do Limite (TCL) e a aproximação da distribuição binomial pela distribuição normal, sabemos que para uma proporção $\widehat{p}_j$ obtida a partir de uma amostra de tamanho $n$,
$$
\widehat{p}_j \sim \text{N}\left(p_j, \frac{p_j (1 - p_j)}{n} \right).
$$
Também sabemos via Método Delta que se uma sequência de variáveis aleatórias $X_1, \ldots, X_n$ satisfaz 
$
\sqrt{n}[X_n - \mu] \xrightarrow{d} \text{N}(0, \sigma^{2}),
$
onde $\mathbb{E}(\bar{X}) = \mu \in \mathbb{R}$ e $\text{Var}(\bar{X}) = \sigma^{2} \in \mathbb{R}^{+}_{*}$, então para qualquer função derivável $g(\cdot)$, tal que $g'(\mu) \ne 0$,
$$
\begin{align}
\sqrt{n}[g(\widehat{p}_j) - g(p_j)] \xrightarrow{d} \text{N}\left(0, \frac{p_j (1 - p_j)}{n} [g'(p_j)]^2 \right).
\end{align}
$$
O resultado acima nos ajuda a encontrar a distribuição de $\log(p_j)$, a qual é dada por
$$
\log(\widehat{p}_j) \sim \text{N}\left(\log(p_j), \frac{(1 - p_j)}{n p_j}\right).
$$
Uma vez que o PSI é uma função de diferenças entre proporções e o logaritmo neperiano da razão entre as mesmas, precisamos encontrar as distribuições assintóticas de $p_j - q_j$ e $\log(p_j)  - \log(q_j)$. Uma vez que todas variáveis aleatórias são normalmente distribuídas com diferentes médias e variâncias,
$$
\widehat{p}_j - \widehat{q}_j \sim \text{N}\left( p_j - q_j, \frac{p_j(1-p_j)}{n} + \frac{q_j(1-q_j)}{m} \right)
$$
e 
$$
\log(\widehat{p}_j) - \log(\widehat{q}_j) \sim \text{N}\left( \log(p_j) - \log(q_j), \frac{1-p_j}{np_j} + \frac{1-q_j}{mp_j} \right).
$$

Agora que sabemos as distribuições assintóticas de $p_j - q_j$ e $\log(p_j)  - \log(q_j)$, podemos nos concentrar em encontrar a esperança e a variância do PSI. A ideia é construirmos um intervalo de confiança com o objetivo de encontrar tamanhos 'ótimos' de amostras $n$ e $m$.

Defina $X_j = p_j - q_j$ e $Y_j = \log(p_j) - \log(q_j)$. Assumindo que $p$ e $q$ são independentes, então $X_j$ e $Y_j$ são independentes. Com isso,
$$
\begin{align}
\mathbb{E}(X_jY_j) & = E(X_j) E(Y_j) \nonumber \\
               & = (p_j - q_j) [\log(p_j) - \log(q_j)]. \nonumber
\end{align}
$$
e
$$
\begin{align}
\text{Var}(X_jY_j) & = \mathbb{E}([X_jY_j]^2) - [\mathbb{E}(X_jY_j)]^2 \nonumber \\
                 & = \mathbb{E}([X_jY_j]^2) - [\mathbb{E}(X_j)\mathbb{E}(Y_j))]^2 \nonumber \\
                 & = \mathbb{E}(X_j^2)\mathbb{E}(Y_j^2) - [\mathbb{E}(X_j)\mathbb{E}(Y_j)]^2 \nonumber \\
                 & = (\sigma^{2}_{xj} + \mu_{xj}^{2}) (\sigma^{2}_{yj} + \mu_{yj}^{2}) - \mu_{xj}^{2} \mu_{yj}^{2} \nonumber \\
                 & = \sigma^{2}_{xj} \sigma^{2}_{yj} + \sigma^{2}_{xj} \mu_{yj}^{2} + \mu_{xj}^{2} \sigma^{2}_{yj}, \nonumber

\end{align}
$$
onde 
$$
\mu_{xj} = p_j - q_j,
$$
$$
\mu_{yj} = \log(p_j/q_j),
$$, 
$$
\sigma^{2}_{xj} = \frac{p_j(1-p_j)}{n} + \frac{q_j(1-q_j)}{m}
$$ 
e
$$
\sigma^{2}_{yj} = \frac{1-p_j}{np_j} + \frac{1-q_j}{mq_j}.
$$
Definindo $\text{Var}(\widehat{\text{PSI}}) = \text{Var}(\sum_{j=1}^{k} X_jY_j) = \sum_{j=1}^{k} \text{Var}(X_jY_j)$, o intervalo de confiança para o PSI é dado por
$$
\widehat{\text{PSI}} \pm Z_{1 - \alpha/2} \times \sqrt{\text{Var}(\widehat{\text{PSI}})}.
$$
Assumindo que $n = m$ e um erro amostral $\epsilon > 0$, temos que o tamanho amostral ótimo é 
$$
n = \frac{Z_{1 - \alpha/2}^{2} \times n \text{Var}(\widehat{\text{PSI}})}{\epsilon^{2}}.
$$

## Visualiza dependência entre $X_j$ e $Y_j$

```python
def diff(p, q):
    return p - q

def ln_diff(p, q):
    return np.log(p) - np.log(q)

def psi_j(p, q):
    return (p - q) * np.log(p / q)

# Create sequences from 0 to 1 with step 0.01
seq_p = np.arange(0, 1.01, 0.01)
seq_q = np.arange(0, 1.01, 0.01)

# Create a grid of all combinations
p_values, q_values = np.meshgrid(seq_p, seq_q)
seq_f = pd.DataFrame({'p': p_values.ravel(), 'q': q_values.ravel()})

# Calculate the metrics
seq_f['diff'] = diff(seq_f['p'], seq_f['q'])
seq_f['ln_diff'] = ln_diff(seq_f['p'], seq_f['q'])
seq_f['psi'] = psi_j(seq_f['p'], seq_f['q'])

# Filter out infinite values (where log was taken of 0)
seq_f = seq_f[np.abs(seq_f['ln_diff']) != np.inf]

(
    ggplot(seq_f, aes(x='diff', y='ln_diff')) +
    geom_point(fill='') + 
    labs(x='p - q', y='log(p) - log(q)', title='Difference vs Log Difference') +
    theme_minimal() +
    theme(figure_size=(8, 6),
          plot_title=element_text(hjust=0.5, size=16),
          axis_title=element_text(size=14),
          axis_text=element_text(size=12),
          legend_title=element_text(size=14),
          legend_text=element_text(size=12))   
)
```

```python
def fx_j(p, q, m=1, n=1):
    mux = p - q
    muy = np.log(p) - np.log(q)
    s2x = p * (1 - p) / n + q * (1 - q) / m
    s2y = (1 - p) / (p * n) + (1 - q) / (q * m)
    var_xy = (s2x + mux**2) * (s2y + muy**2) - mux**2 * muy**2
    return var_xy

def psi_j(p, q):
    return (p - q) * np.log(p / q)

seq_p = np.arange(0, 1, 0.001)
seq_q = np.arange(0, 1, 0.001)

p_grid, q_grid = np.meshgrid(seq_p, seq_q)
p_flat = p_grid.flatten()
q_flat = q_grid.flatten()

var = np.array([fx_j(p, q) for p, q in zip(p_flat, q_flat)])
psi = np.array([psi_j(p, q) for p, q in zip(p_flat, q_flat)])

seq_f = pd.DataFrame({'p': p_flat, 'q': q_flat, 'var': var, 'psi': psi})
seq_f = seq_f.sort_values('p').drop_duplicates('var')
```

```python
print(
    ggplot(seq_f, aes(x='p', y='q', fill='psi')) +
    geom_tile() +
    scale_fill_cmap('turbo') +
    theme_minimal() +
    labs(title='PSI heatmap', x='p', y='q', fill='PSI') +
    theme(figure_size=(8, 6),
          plot_title=element_text(hjust=0.5, size=16),
          axis_title=element_text(size=14),
          axis_text=element_text(size=12),
          legend_title=element_text(size=14),
          legend_text=element_text(size=12))
)

print(
    ggplot(seq_f[seq_f.psi <= 1], aes(x='p', y='q', fill='psi')) +
    geom_tile() +
    scale_fill_cmap('turbo') +
    theme_minimal() +
    labs(title='PSI <= 1', x='p', y='q', fill='PSI') +
    theme(figure_size=(8, 6),
          plot_title=element_text(hjust=0.5, size=16),
          axis_title=element_text(size=14),
          axis_text=element_text(size=12),
          legend_title=element_text(size=14),
          legend_text=element_text(size=12))
)

print(
    ggplot(seq_f[seq_f.psi <= 0.25], aes(x='p', y='q', fill='psi')) +
    geom_tile() +
    scale_fill_cmap('turbo') +
    theme_minimal() +
    labs(title='PSI <= 0.25', x='p', y='q', fill='PSI') +
    theme(figure_size=(8, 6),
          plot_title=element_text(hjust=0.5, size=16),
          axis_title=element_text(size=14),
          axis_text=element_text(size=12),
          legend_title=element_text(size=14),
          legend_text=element_text(size=12))
)

print(
    ggplot(seq_f[seq_f.psi <= 0.1], aes(x='p', y='q', fill='psi')) +
    geom_tile() +
    scale_fill_cmap('turbo') +
    theme_minimal() +
    labs(title='PSI <= 0.1', x='p', y='q', fill='PSI') +
    theme(figure_size=(8, 6),
          plot_title=element_text(hjust=0.5, size=16),
          axis_title=element_text(size=14),
          axis_text=element_text(size=12),
          legend_title=element_text(size=14),
          legend_text=element_text(size=12))
)

print(
    ggplot(seq_f, aes(x='p', y='q', fill='var')) +
    geom_tile() +
    scale_fill_cmap('turbo') +
    theme_minimal() +
    labs(title='Variance of PSI heatmap', x='p', y='q', fill='Var(PSI)') +
    theme(figure_size=(8, 6),
          plot_title=element_text(hjust=0.5, size=16),
          axis_title=element_text(size=14),
          axis_text=element_text(size=12),
          legend_title=element_text(size=14),
          legend_text=element_text(size=12))
)

print(
    ggplot(seq_f[seq_f.psi <= 0.1], aes(x='p', y='q', fill='var')) +
    geom_tile() +
    scale_fill_cmap('turbo') +
    theme_minimal() +
    labs(title='Variance of PSI, where PSI <= 0.1', x='p', y='q', fill='Var(PSI)') +
    theme(figure_size=(8, 6),
          plot_title=element_text(hjust=0.5, size=16),
          axis_title=element_text(size=14),
          axis_text=element_text(size=12),
          legend_title=element_text(size=14),
          legend_text=element_text(size=12))
)

print(
    ggplot(seq_f[(seq_f.psi <= 0.1) & (seq_f.p <= 0.05)], aes(x='p', y='q', fill='var')) +
    geom_tile() +
    scale_fill_cmap('turbo') +
    theme_minimal() +
    labs(title='Variance of PSI, where PSI <= 0.1 & p <= 0.05', x='p', y='q', fill='Var(PSI)') +
    theme(figure_size=(8, 6),
          plot_title=element_text(hjust=0.5, size=16),
          axis_title=element_text(size=14),
          axis_text=element_text(size=12),
          legend_title=element_text(size=14),
          legend_text=element_text(size=12))
)
```
